```python
import pyodbc

# Specifying the ODBC driver, server name, database, etc. directly
cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER=qa2db3;DATABASE=equestplus;Trusted_Connection=True')

cursor = cnxn.cursor()

cursor.execute("SELECT Distinct OptionName FROM eQuestScreenOptions with (NOLOCK) WHERE ScreenID='BGRPT' ORDER BY OptionName")

reportRuleOptionNames = cursor.fetchall()

reportRuleOptionNames
```
