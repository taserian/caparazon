```python
import pyodbc
import functools

class Search4Item:
    """A generic search item"""
    def __init__(self, item_type, qualified_name, search_term):
        self._item_type = item_type
        self._qualified_name = qualified_name
        self._search_term = search_term
        
    def search4(self):
        return self._search_term
    
    def __str__(self):
        return "({}, {}, {})".format(self._item_type, self._qualified_name, self._search_term)
    
## 

class ReportRule(Search4Item):
    """holder for Report Rule (also called Screen Option)"""
    pass
    
class CustomerOption(Search4Item):
    pass
     
class SQLCallableName(Search4Item):
    pass

# Specifying the ODBC driver, server name, database, etc. directly
cnxn = pyodbc.connect('DRIVER={SQL Server};SERVER=qa2db3;DATABASE=equestplus;Trusted_Connection=True')
cursor = cnxn.cursor()
# 
cursor.execute("SELECT Distinct 'RR', 'ReportRule: '+ ScreenId + '.' + OptionName, OptionName FROM eQuestScreenOptions with (NOLOCK) ORDER BY OptionName")
report_rule_option_names = [ReportRule(x, y, z) for (x, y, z) in cursor.fetchall()]

cursor.execute("SELECT Distinct 'CO', 'CustomerOption: ' + CustOptionName, CustOptionName FROM CustomerOptions with (NOLOCK)")
customer_option_names = [CustomerOption(x, y, z) for (x, y, z) in cursor.fetchall()]

cursor.execute("""
select (CASE routine_type WHEN 'PROCEDURE' THEN 'SQLSP' WHEN 'FUNCTION' THEN 'SQLF' END),
       (CASE routine_type WHEN 'PROCEDURE' THEN 'SQL Stored Procedure: ' WHEN 'FUNCTION' THEN 'SQL Function: ' END) 
           + '[' + ROUTINE_CATALOG + '].[' + ROUTINE_SCHEMA + '].[' + ROUTINE_NAME + ']',
       ROUTINE_NAME 
from equestplus.information_schema.routines with (nolock) 
where routine_type IN ('PROCEDURE', 'FUNCTION')
order by ROUTINE_CATALOG, ROUTINE_SCHEMA, ROUTINE_NAME 
""")
sql_item_names = [SQLCallableName(x, y, z) for (x, y, z) in cursor.fetchall()]

search4_items = report_rule_option_names + customer_option_names + sql_item_names
for x in search4_items:
    print(x)
```
